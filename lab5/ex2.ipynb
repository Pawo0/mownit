# Kompresja obrazu
1. Przygotuj przykładowe zdjęcie z skali szarości o rozmiarze co najmniej 512 × 512
pikseli (np. Lenna image)
2. Oblicz SVD macierzy pikseli I, a następnie dokonaj przybliżenia tej macierzy za pomocą low rank approximation (k pierwszych wartości osobliwych) uzyskując kompresję obrazu wejściowego.
3. Porównaj obraz wynikowy z obrazem źródłowym dla różnych wartości k (przedstawiając różnicę pomiędzy nimi w postaci zdjęcia oraz rysując wykres zależności
||I − Ia|| od k).

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import scipy as sp
```

```python
# Wczytanie obrazu
img = Image.open('lena_gray.gif').convert('L')

# calculate SVD
U, S, Vt = np.linalg.svd(img, full_matrices=False)
# manual version


# Wczytanie obrazu
img = Image.open('lena_gray.gif').convert('L')
img = np.array(img)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()

# kompresja obrazu
K = [1, 5, 10, 20, 50, 100, 200]
for k in K:
    IA = np.dot(U[:, :k] * S[:k], Vt[:k, :])

    # Normalizacja wartości do zakresu [0, 255]
    IA = np.clip(IA, 0, 255).astype(np.uint8)

    plt.imshow(1 - IA, cmap='gray')
    plt.axis('off')
    plt.title(f'k = {k}')
    plt.show()


```
