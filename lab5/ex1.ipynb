```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# rownanie parametryczne sfery jednostkowej
def sphere(radius=1, num_points=1000):
    phi = np.linspace(0, np.pi, num_points)
    theta = np.linspace(0, 2 * np.pi, num_points)
    x = radius * np.outer(np.sin(phi), np.cos(theta))
    y = radius * np.outer(np.sin(phi), np.sin(theta))
    z = radius * np.outer(np.cos(phi), np.ones(num_points))
    return x, y, z

# wizualizowanie sfery
def plot_sphere(x, y, z, ax):
    ax.plot_surface(x, y, z, color='b', alpha=0.5)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Sfera jednostkowa')
    plt.show()

```

```python
# wyswietlenie sfery jednostkowej
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x, y, z = sphere()
plot_sphere(x, y, z, ax)
```
